```python
import matplotlib.pyplot as plt
%matplotlib inline

days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[], color='c', label='sleeping', linewidth=5)
plt.plot([],[], color='b', label='eating', linewidth=5)
plt.plot([],[], color='m', label='working', linewidth=5)
plt.plot([],[], color='k', label='playinging', linewidth=5)
plt.stackplot(days,sleeping,eating,working,playing, colors= ['c','b','m','k'])
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```
